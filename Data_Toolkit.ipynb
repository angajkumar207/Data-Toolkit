4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

```python
import seaborn as sns
df = pd.DataFrame({'col1': [1, 2, 3, 4, 5], 'col2': [10, 20, 30, 40, 50]})
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

8. Create a line plot with multiple lines using Matplotlib.

```python
# multiple
import matplotlib.pyplot as plt
import numpy as np

x_data = np.array([1, 2, 3, 4, 5])
y_data1 = np.array([2, 4, 6, 8, 10])
y_data2 = np.array([1, 3, 5, 7, 9])
y_data3 = np.array([0, 5, 10, 15, 20])

plt.plot(x_data, y_data1, label='Line 1')
plt.plot(x_data, y_data2, label='Line 2')
plt.plot(x_data, y_data3, label='Line 3')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Multiple Lines')
plt.legend()
plt.legend()
plt.show()
```

9.  Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
